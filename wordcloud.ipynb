```python
from wordcloud import WordCloud, STOPWORDS
from GoogleNews import GoogleNews
```

```python
news = GoogleNews(lang='en', region='US', period='30d')
```

```python
news.get_news(key='Russia')
```

```python
results = news.results()
```

```python
topics = [result.get('title') for result in results]
```

```python
topics[:5]
```

```python
topics_single_str = '\n'.join(topics)
```

```python
stopwords = STOPWORDS
```

```python
stopwords.update(('Russian', 'Now', 'Will', 'will', 'See', 'Already', 'Sees', 'U', 'S', 'say', 'says'))
```

```python
wordcloud = WordCloud(stopwords=stopwords, background_color='white', width=600, height=400, max_words=50, min_font_size=8, max_font_size=None).generate(topics_single_str)
```

```python
img = wordcloud.to_image()
```
